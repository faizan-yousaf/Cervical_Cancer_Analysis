## Meta-Data (About Dataset)

* `Cervical Cancer Risk Factors for Biopsy`:


This dataset, obtained from the UCI Repository, explores various **risk factors** associated with **cervical cancer** leading to **biopsy examinations**. Each year, about **11,000 new cases** of invasive cervical cancer are diagnosed in the U.S. Despite a steady decline in new cases over recent decades, cervical cancer remains a significant health issue, causing approximately **4,000 deaths** annually in the U.S. and around **300,000 worldwide**.

* `Age`:

Cervical cancer diagnoses predominantly occur in women aged **35 to 54**, with about **50%** of cases falling within this range. The **median age** for diagnosis is **48 years**. While it's rare for women under **20** to develop cervical cancer, many young women can become infected with multiple types of **human papilloma virus (HPV)**. Without regular screenings, these women may be at high risk for localized or invasive cancer later in life.

* `Socioeconomic and Ethnic Factors`:

Although cervical cancer rates have decreased among **Caucasian** and **African-American** women, African-American women still face higher prevalence and death rates, which are **twice** as high as their Caucasian counterparts. **Hispanic American** women also experience more than **double** the risk compared to Caucasian women, largely due to lower screening rates linked to socioeconomic factors. High **poverty levels**, lack of **health insurance**, limited **transportation**, and **language barriers** contribute to disparities in access to screening services.

* `High Sexual Activity`:

The primary risk factor for cervical cancer is **HPV**, primarily transmitted through **sexual activity**. Women with multiple sexual partners or who become sexually active at a young age are at increased risk. A history of **sexually transmitted infections**, such as **chlamydia**, has also been associated with higher cervical cancer risk.

* `Family History`:

Women with a **first-degree relative** (mother or sister) diagnosed with cervical cancer have a higher risk of developing the disease themselves.

* `Use of Oral Contraceptives`:

Long-term use of **oral contraceptives** (more than **5-10 years**) has been linked to an increased risk of cervical cancer, particularly due to a potential higher likelihood of HPV infection. The exact mechanisms behind this association remain unclear.

* `Having Many Children`:

Research indicates that having **multiple children** can elevate the risk of cervical cancer, especially among women infected with HPV.

* `Smoking`:

**Smoking** has been linked to an increased risk of precancerous changes in the cervix and progression to invasive cervical cancer, particularly in women with HPV infections.

* `Immunosuppression`:

Women with compromised **immune systems**, such as those with **HIV/AIDS**, are more vulnerable to acquiring HPV and may experience faster progression from cervical precancer to invasive cancer.

* `Diethylstilbestrol (DES)`:

Between **1938 and 1971**, the drug **diethylstilbestrol (DES)** was prescribed to pregnant women to prevent miscarriages. Daughters of these women face a heightened risk for cervical cancer, although DES is no longer prescribed.

---

This comprehensive overview of cervical cancer risk factors underscores the importance of **awareness**, **screening**, and **prevention strategies**. 

## Context

This dataset contains **835 records** and **36 columns**. It focuses on factors that may influence the likelihood of cervical cancer. The variables capture patient demographics, lifestyle factors, sexual health, and results from various medical tests.

### Content: 

#### Column Descriptions**

- **`Age`**:  
   The age of the patient in years (integer).
  
- **`Number of sexual partners`**:  
   Number of lifetime sexual partners (float).
  
- **`First sexual intercourse`**:  
   Age at first sexual intercourse (float).
  
- **`Num of pregnancies`**:  
   Number of pregnancies the patient has had (float).

- **`Smokes`**:  
   Binary indicator of whether the patient smokes (`0` = No, `1` = Yes).
  
- **`Smokes (years)`**:  
   Number of years the patient has been smoking (float).
  
- **`Smokes (packs/year)`**:  
   Average number of packs smoked per year (float).

- **`Hormonal Contraceptives`**:  
   Binary indicator for the use of hormonal contraceptives (`0` = No, `1` = Yes).
  
- **`Hormonal Contraceptives (years)`**:  
   Duration of hormonal contraceptive use in years (float).

- **`IUD`**:  
   Binary indicator for the use of an intrauterine device (`0` = No, `1` = Yes).
  
- **`IUD (years)`**:  
   Duration of IUD use in years (float).

- **`STDs`**:  
   Binary indicator for whether the patient has had sexually transmitted diseases (STDs) (`0` = No, `1` = Yes).

- **`STDs (number)`**:  
   Number of distinct STDs the patient has had (float).

- **STD-specific columns**:
   - **`STDs:condylomatosis`**: Presence of condylomatosis (genital warts) (`0` = No, `1` = Yes).
   - **`STDs:cervical condylomatosis`**: Cervical condylomatosis presence (`0` = No, `1` = Yes).
   - **`STDs:vaginal condylomatosis`**: Vaginal condylomatosis (`0` = No, `1` = Yes).
   - **`STDs:vulvo-perineal condylomatosis`**: Vulvo-perineal condylomatosis (`0` = No, `1` = Yes).
   - **`STDs:syphilis`**: Syphilis diagnosis (`0` = No, `1` = Yes).
   - **`STDs:pelvic inflammatory disease`**: Pelvic inflammatory disease (`0` = No, `1` = Yes).
   - **`STDs:genital herpes`**: Genital herpes (`0` = No, `1` = Yes).
   - **`STDs:molluscum contagiosum`**: Molluscum contagiosum (`0` = No, `1` = Yes).
   - **`STDs:AIDS`**: AIDS diagnosis (`0` = No, `1` = Yes).
   - **`STDs:HIV`**: HIV diagnosis (`0` = No, `1` = Yes).
   - **`STDs:Hepatitis B`**: Hepatitis B diagnosis (`0` = No, `1` = Yes).
   - **`STDs:HPV`**: Human papillomavirus (HPV) diagnosis (`0` = No, `1` = Yes).

- **`STDs: Number of diagnosis`**:  
   Total number of STD diagnoses (integer).

- **`STDs: Time since first diagnosis`**:  
   Time (in years) since the patient‚Äôs first STD diagnosis (float).

- **`STDs: Time since last diagnosis`**:  
   Time (in years) since the patient‚Äôs most recent STD diagnosis (float).

#### **Medical Diagnosis Columns**
  
- **`Dx:Cancer`**:  
   Binary indicator for a cancer diagnosis (`0` = No, `1` = Yes).

- **`Dx:CIN`**:  
   Binary indicator for Cervical Intraepithelial Neoplasia (CIN) diagnosis (`0` = No, `1` = Yes).

- **`Dx:HPV`**:  
   Binary indicator for HPV diagnosis (`0` = No, `1` = Yes).

- **`Dx`**:  
   General diagnosis of any health condition (`0` = No, `1` = Yes).

#### **Test Results**

- **`Hinselmann`**:  
   Result of the Hinselmann test for cervical cancer detection (`0` = Negative, `1` = Positive).

- **`Schiller`**:  
   Result of the Schiller test (`0` = Negative, `1` = Positive).

- **`Citology`**:  
   Result of the cytological test (`0` = Negative, `1` = Positive).

- **`Biopsy`**:  
   Result of a cervical biopsy (`0` = Negative, `1` = Positive).


This dataset captures a wide range of **risk factors for cervical cancer** and **related medical conditions**, including lifestyle factors, sexual health history, and results of diagnostic tests. It may be used for research in **predicting cervical cancer risk** based on these factors.
